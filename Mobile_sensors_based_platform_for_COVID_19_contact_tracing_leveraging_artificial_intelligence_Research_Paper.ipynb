# Random Forest Classifier

```python
import pandas as pd
col_list = ['rssi','ax','ay','az','gx','gy','gz','speed','behavior']
data = pd.read_csv('/content/drive/MyDrive/NewPaperWork_vvvvvvvvvvvvv2/finaldataset.csv', usecols=col_list)
data = data.dropna()
```

```python
from sklearn.model_selection import train_test_split
X = data[['rssi','ax','ay','az','gx','gy','gz','speed']]
y = data[['behavior']]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
from sklearn.metrics import confusion_matrix
import sklearn.metrics as metrics
from sklearn.metrics import mean_absolute_error

model = RandomForestClassifier(n_estimators=750,
                               criterion='entropy',
                               max_depth= 68
                               )
model.fit(X_train, y_train.values.ravel())

# make predictions
y_pred = model.predict(X_test)
# evaluate predictions
```

```python
acc = accuracy_score(y_test, y_pred)
print('Accuracy: %.3f' % (100*acc)+'%')
MAE = mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error: %.3f' % (MAE))
explained_variance=metrics.explained_variance_score(y_test, y_pred)

print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('Coefficient of Determination %0.3f' % explained_variance)
```
